### Explanation of Research & Testing

We performed 3 experiments on the dataset, these are:
1. Features predicting salary class
2. Distinguishing factors between Data Analysts/ Data Scientists/ Data Engineers roles
3. Distinguishing factors between Junior/ Senior position


**Data Overview** 

Searched keywords: data analyst, data scientist, business analyst, business intelligence, data engineer and analytics. 

Scraped from mycareersfuture.sg - **914 rows** and **7 columns**, after cleaning
feature engineering added **6 columns**. 

Columns: 'company_name', 'employment_type', 'industry', 'job_description',
       'job_title', 'requirements', 'salary_max', 'salary_min', 'seniority',
       'salary_avg', 'job_title_clean', 'seniority_clean',
       'salary_class'

* **Experiment 1** - Target feature was salary class, treshold deterimined by IQR. 
       * High Salary is > IQR 75, Low Salary is < IQR 25
       * high salary treshold:  9075, low salary treshold:  5500
       * Baseline is 'mid' class between high and low salary threshold at ~50% of the entire dataset. 

* **Experiment 2** - performed against subsets of the data, where Job_title matched Data Analyst, Data Scientist or Data Engineer. Target feature was Job Title. Number of Samples: 
        * Data Engineer     138 
        * Data Analyst      120 
        * Data Scientist     80

* **Experiment 3** - performed against subsets of the data, where seniority was further classified into Junior and Senior. Number of samples:    
        * Junior    240
        * Senior    228

Data scraping and cleaning in separate notebook

**Problem with job description**: from EDA it seems that many job descriptions were identical to each other. Meaning that different salary levels can have the same job description, which can confuse our models and increase the risk of misclassification.

**Hypothesis:**
- Prediction will perform better without job description feature

**Tests per experiment**
- Test on Job Title, Industry Seniority, Salary class
- Test on JD
- Test on Requirements
- Test on all Features

**Models**
- Logistic Regression
- Stochastic Gradient Classifier
- Random Forest Classifier

**Evaluation Metric**
- F1
- Confusion Matrix
- Accuracy Score


## RESULTS 

**Experiment Results**

1. On Salary Prediction:
We tried predicting the class of salary with a few combination of all features. Best performance was delivered by test against Job Requirements alone. Our best performing test confirm that: 
        High salary jobs require softskills, such as relationship and management. Whereas Mid to Low salary look 
        more into qualifications such as degree and achievements. 
        
2. On Job Functions: Data Analyst/ Data Scientist and Data Engineer. Result showed that skills required from a Data Analyst vary from Data Engineer or Data Scientist: 
        * Analytics, Reporting, Financial or Business background is useful for Data Analysts, whereas
        * Software engineering, technical chops, and at least a bachelor degree in computer science are useful for Data Engineer, whereas
        * Modeling, statistical knowledge and research are essential for Data Scientist.
        
3. On Junior vs Senior Positions: Unlike the first two experiments where Job Description didn't contain strong factors predicting target variable, job description contained the most important features determining the seniority of positions. Test for keywords returned inconclusive however, with very general skills from senior position such as management skills and year of experience.

**Model Results**

Best performing model was logistic regression, which is consistent amongst all the tests conducted on Experiment 1. We used only Logistic Regression for Experiment 2 and 3.
